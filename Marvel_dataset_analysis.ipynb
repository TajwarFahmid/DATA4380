```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_bar_chart(df, column_name):
    """
    Generate a bar chart for a specified column of a DataFrame.

    Parameters:
        df (DataFrame): The DataFrame containing the data.
        column_name (str): The name of the column to plot.
    """
    # Aggregate data by the specified column
    agg_data = df[column_name].value_counts()

    # Plotting
    plt.figure(figsize=(10, 8))
    agg_data.plot(kind='bar', color='skyblue')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Bar Chart for {column_name}')
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed
    plt.tight_layout()  # Adjust layout to prevent clipping of labels
    plt.show()


#Using this function we can plot bar graphs for any column in the dataset
df = pd.read_csv('/content/marvel_box_office.csv')
plot_bar_chart(df, 'Release Month')
```

```python
# Now we can generate a function to find the mean and median of columns such as revenue generated, box office income etc.
import pandas as pd

def calculate_mean_and_median(df, column_names):
    """
    Calculate the mean and median of specified columns in a DataFrame.

    Parameters:
        df (DataFrame): The DataFrame containing the data.
        column_names (list): A list of column names for which to calculate the mean and median.

    Returns:
        mean_and_median_values (DataFrame): A DataFrame containing the mean and median values for each specified column.
    """
    mean_values = df[column_names].mean()
    median_values = df[column_names].median()

    mean_and_median_values = pd.concat([mean_values, median_values], axis=1)
    mean_and_median_values.columns = ['Mean', 'Median']

    return mean_and_median_values

df = pd.read_csv('/content/marvel_box_office.csv')

columns_to_analyze = ['Domestic Box Office', 'International Box Office', 'Budget','IMDb Score']

results = calculate_mean_and_median(df, columns_to_analyze)
print("Mean and Median values:")
print(results)
```

```python
# We can generate a pie chart to show how the ownership of the marvel movies are divided
def generate_pie_chart(df, column_name):

        # Count the occurrences of each category in the specified column
    value_counts = df[column_name].value_counts()

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'{column_name}')
    plt.axis('equal')
    plt.show()

df = pd.read_csv('/content/marvel_box_office.csv')

# Specify the column for which you want to generate the pie chart
column_to_plot = 'Ownership'

# Generate the pie chart for the specified column
generate_pie_chart(df, column_to_plot)
```
